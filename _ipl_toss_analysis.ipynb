```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Bat first\n(after toss win)', 'Chase\n(after toss win)']
win_rates = [48, 53]
colors = ['#d9534f', '#1D9E75']

fig, ax = plt.subplots(figsize=(10.8, 10.8), dpi=100)
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

# Bars
bars = ax.bar(categories, win_rates, color=colors,
              width=0.45, zorder=3, edgecolor='white', linewidth=0)

for bar, val in zip(bars, win_rates):
    ax.text(bar.get_x() + bar.get_width()/2,
            bar.get_height() + 0.8,
            f'{val}%', ha='center', va='bottom',
            fontsize=32, fontweight='bold',
            color=bar.get_facecolor())

ax.annotate('', xy=(1, 55.5), xytext=(0, 50.5),
            arrowprops=dict(arrowstyle='->', color='#888',
                           lw=1.5, connectionstyle='arc3,rad=0.2'))
ax.text(0.5, 54.5, '+5% edge', ha='center', fontsize=14,
        color='#555', style='italic')

ax.set_ylim(0, 75)

ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#f0f0f0', linewidth=0.8)
ax.set_yticks([0, 25, 50, 75])
ax.tick_params(axis='y', labelsize=12, colors='#bbb')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#e0e0e0')
ax.tick_params(axis='x', labelsize=18, pad=12)

ax.set_title('Chasing wins more than batting first in IPL',
             fontsize=22, fontweight='bold', pad=28,
             color='#1a1a1a')

# subtitle moved lower so it doesn't overlap title
fig.text(0.5, 0.86,
         '1,000+ matches analysed · IPL 2008–2025',
         ha='center', fontsize=13, color='#888')


fig.text(0.95, 0.02, 'linkedin.com/in/nikita-jadhav0903',
         ha='right', fontsize=11, color='#bbb')

plt.tight_layout(rect=[0, 0.04, 1, 0.92])
plt.savefig('ipl_toss_insight.png', dpi=100,
            bbox_inches='tight', facecolor='white')
plt.show()
print("Saved as ipl_toss_insight.png")
```
